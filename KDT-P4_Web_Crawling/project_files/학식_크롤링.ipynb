```python
from  bs4 import BeautifulSoup
from urllib.request import urlopen 
from datetime import datetime

center_url = 'https://coop.knu.ac.kr/sub03/sub01_01.html?shop_sqno=35'
bocgi_url = 'https://coop.knu.ac.kr/sub03/sub01_01.html?shop_sqno=36'
gong_url = 'https://coop.knu.ac.kr/sub03/sub01_01.html?shop_sqno=85'


```

```python
ordinal_menu = ['','일반돈가스★￦ 3,000','통등심돈가스★￦ 3,500','통등심\n치즈돈가스★￦ 3,800','고구마치즈돈가스★￦ 3,800','왕돈가스★￦ 5,000','왕돈가스★￦ 4,500','치즈왕돈가스★￦ 5,000','비건스테이크￦ 4,500','라면￦ 2,800','어묵우동￦ 3,500','어묵우동밥￦ 3,800']
    
```

```python
def get_center_menu():
    html = urlopen(center_url)
    soup = BeautifulSoup(html, 'html.parser')

    day_night =  soup.select('.menu_im')
    lunch_menu = day_night[datetime.now().weekday()]
    dduck_menu = day_night[datetime.now().weekday()+6]
    dinner_menu = day_night[datetime.now().weekday()+12]


    menus =[lunch_menu, dduck_menu,dinner_menu]
    center_menu = []
    for men in menus:
        menu = ''
        for one_menu in men.select('li'):
            if not one_menu.text in ordinal_menu:
                menu = menu +'\n'+ (one_menu.text)
        if menu != '':
            center_menu.append(menu)
    return center_menu
```

```python
def get_gong_menu():
    html = urlopen(gong_url)
    soup = BeautifulSoup(html, 'html.parser')
    ay_night = soup.select('.menu_im')

    day_night = soup.select('.menu_im')
    lunch_menu = day_night[datetime.now().weekday()]
    dinner_menu = day_night[datetime.now().weekday()+6]

    menus = [lunch_menu, dinner_menu]
    gong_menu = []
    for men in menus:
        menu = ''
        for one_menu in men.select('li'):
            menu = menu + '\n' + (one_menu.text)
        if menu != '':
            gong_menu.append(menu)

    return gong_menu
```

```python
import pandas as pd


def get_all_menu(kind):
    try:
        center= get_center_menu()
    except:
        center = ['','','']
    bocgi = get_bocgi_menu()
    gong = get_gong_menu()

    menu_today = pd.DataFrame({'밥센터': [center[0], center[1]+center[2]],
                               '복지관' : [bocgi[0],''],
                               '공식당' : [gong[0],gong[1]]},
                              index = ['점심', '저녁']
                             )
    print(datetime.today())
    print('--------------',kind,'------------------\n\n')
    for c, m in enumerate(menu_today.loc[kind]):
        print(menu_today.columns[c]+'--------------------',m,'\n--------------------------\n')
get_all_menu('점심')
get_all_menu('저녁')
```
