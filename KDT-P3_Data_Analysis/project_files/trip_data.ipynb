## 연도별 평균(성별) DF화

```python
a = [tnDF_1_idx['201801':'201812']['남성'].mean(),
     tnDF_1_idx['201901':'201912']['남성'].mean(),
    tnDF_1_idx['202001':'202012']['남성'].mean(),
    tnDF_1_idx['202101':'202112']['남성'].mean(),
    tnDF_1_idx['202201':]['남성'].mean()]
b = [tnDF_1_idx['201801':'201812']['여성'].mean(),
     tnDF_1_idx['201901':'201912']['여성'].mean(),
    tnDF_1_idx['202001':'202012']['여성'].mean(),
    tnDF_1_idx['202101':'202112']['여성'].mean(),
    tnDF_1_idx['202201':]['여성'].mean()]
year = ['2018', '2019', '2020', '2021', '2022']

tnDF_year_mean = pd.DataFrame({'male' : a, 'female' : b}, index = year)
tnDF_year_mean
```

## 연도별 성별(평균) 막대그래프

```python
plt.figure(figsize=(18, 13))
tnDF_year_mean.plot(kind='bar')

#.bar(x=tnDF_year_mean.index, height=tnDF_year_mean.male, tnDF_year_mean.female)
# plt.bar(x=tnDF_year_mean.index, height=tnDF_year_mean.female)
plt.xticks(rotation = 45, fontsize=10)
plt.legend(['male', 'female'], fontsize=20)
plt.title('gender(mean)', fontsize=30) # 그래프 타이틀
plt.show
```
