```python
# MODULE 
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
from matplotlib import font_manager as fm, rc
import import_ipynb

# 한글 폰트 설정
font_path='./font/malgun.ttf'
font_name=fm.FontProperties(fname=font_path).get_name()
rc('font',family=font_name)
```

```python
# 파일 import
import travel_expense as TE
import travel_number as TN
```

```python
expenseDF_allyear=TE.expenseDF_allyear
expenseDF_allyear.head()
```

```python
travelNumberDF_allyear=TN.travelNumberDF_allyear
travelNumberDF_allyear.head()
```

```python
expenseDF_allyear=expenseDF_allyear*100000  # 단위:만원
travelNumberDF_allyear=travelNumberDF_allyear*1000
expenseDF_allyear
```

```python
travelNumberDF_allyear
```

```python
# 여행 횟수 당 지출 DF 생성
ex_per_travelDF=np.divide(expenseDF_allyear,travelNumberDF_allyear)

# 컬럼명 변경
ex_per_travelDF.columns=[2018,2019,2020]

# 소숫점 2자리까지, 반올림
ex_per_travelDF=ex_per_travelDF.round(2)

ex_per_travelDF

# (단위: 만원)
```

```python
plt.figure(figsize=(10,8))
line1=plt.plot(ex_per_travelDF.index, ex_per_travelDF[2018],'bo-',ms=5,label='2018년')
line2=plt.plot(ex_per_travelDF.index, ex_per_travelDF[2019],'go-',ms=5,label='2019년')
line3=plt.plot(ex_per_travelDF.index, ex_per_travelDF[2020],'ro-',ms=5,label='2020년')

plt.title('월별 국내여행 횟수당 지출 (2018년~2020년)')
plt.xticks(list(ex_per_travelDF.index))
plt.grid(True)

lines=line1+line2+line3
labels=[l.get_label() for l in lines]
plt.legend(lines, labels, loc='upper right')
```
